### Question 1


The file "https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/files.txt" contains a list of CSV files.

Write a python code which reads all CSV files listed in the above file into a single pandas DataFrame "grades".

Then sort this DataFrame with respect to the "id" column.

```python
files = pd.read_csv("https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/files.txt", header=None)
files
```

```python
grades = None
for i in files[0].values:
    df = pd.read_csv(i)
    grades = pd.concat([grades,df])
grades.sort_values("ID", inplace=True)
grades.reset_index(drop=True, inplace=True)
grades.head()
```

```python
grades = grades.merge(project[["ID","Project"]])
grades.head()
```

### Questions 3 and 4

The file "https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/levels.csv" contains a list of CSV files.

Write a python code which reads all CSV files listed in the above file into a single pandas DataFrame "grades".

```python
grades.set_index(["Group","ID"], inplace=True)
grades.sort_index(inplace = True)
```

```python
grades.head()
```

```python
levels = pd.read_csv("https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/levels.csv")
levels.head()
```

```python
levels_dict = {}

for i in range(levels.shape[0]):
    levels_dict[levels.iloc[i,0]] = levels.iloc[i,1]
```

```python
levels_dict
```

```python
pairs = []

for i in grades.columns:
    pairs.append((levels_dict[i],i))
    
grades_index = pd.MultiIndex.from_tuples(pairs)
grades_index
```

```python
ModIndex = pd.MultiIndex.from_arrays(levels.values.T)
```

```python
grades.columns = grades_index
```

### Question 5

```python
grades.loc[:,["Level 4","Level 5","Level 6","Final Year"]].head()
```

```python
mean_grades = grades.iloc[:,:2]
mean_grades.head()
```

```python
mean_grades[("Annual Grades","Level 4")] = grades.loc[:,"Level 4"].mean(axis = 1).round(1)
mean_grades[("Annual Grades","Level 5")] = grades.loc[:,"Level 5"].mean(axis = 1).round(1)
mean_grades[("Annual Grades","Level 6")] = grades.loc[:,"Level 6"].mean(axis = 1).round(1)
mean_grades[("Annual Grades","Final Year")] = grades.loc[:,("Final Year","Project")].round(1)
mean_grades[("Degree","Final Grade")] = grades.loc[:,["Level 4","Level 5","Level 6","Final Year"]].mean(axis = 1).round(1)
```

```python
classes = pd.cut(mean_grades[("Degree","Final Grade")],[39.50,49.50,59.50,69.50,100], labels=["Third", "Lower Second", "Upper Second", "First"])
mean_grades[("Degree","Class")] = classes
classes.head()
```

```python
mean_grades.head()
```

### Question 6

Compute the number of students in each class

```python
degrees = pd.DataFrame(mean_grades.pivot_table(index = [("Degree","Class")], aggfunc ='size'))
degrees.index.name = "Degree"
degrees.columns = ["Count"]
degrees
```

```python
degrees.to_csv("degrees.csv")
```
