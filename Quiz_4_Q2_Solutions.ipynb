### Solutions

#### Python code

````python
import numpy as np
import pandas as pd
````

````python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
````

````python
# movies

movies = pd.read_csv("https://raw.githubusercontent.com/uqglmn/dhv/main/q4/imdb.csv")[["movie","year","genre","gross_earning"]]

movies = movies.loc[movies["gross_earning"].str.startswith("$")]
movies["gross_earning"] = (movies["gross_earning"].replace("[$M]", "", regex=True).astype(float) * 10**6).astype(int)
movies["year"] = movies["year"].replace("[()I ]", "", regex=True).astype(int)
movies.reset_index(drop=True, inplace=True)

# top_action, top_adventure, top_comedy

cols = ["movie","year","gross_earning"]

df = movies[movies["genre"]=="Action"]
max_indices = df.groupby("year")["gross_earning"].transform(max) == df["gross_earning"]
top_action = df.loc[max_indices][cols].sort_values("year").reset_index(drop=True)

df = movies[movies["genre"]=="Adventure"]
max_indices = df.groupby("year")["gross_earning"].transform(max) == df["gross_earning"]
top_adventure = df.loc[max_indices][cols].sort_values("year").reset_index(drop=True)

df = movies[movies["genre"]=="Comedy"]
max_indices = df.groupby("year")["gross_earning"].transform(max) == df["gross_earning"]
top_comedy = df.loc[max_indices][cols].sort_values("year").reset_index(drop=True)

# top_three_genres

sort_cols = ['genre','gross_earning'] 

df = movies.loc[movies["year"] >= 2000].groupby("genre")["gross_earning"].nlargest(3)
max_genres = df.groupby("genre").sum().sort_values()[-3:].index
max_indexes = df[max_genres].reset_index()["level_1"]
top_three_genres_Q21 = movies.loc[max_indexes,:].sort_values(sort_cols, ascending = [True, False]).reset_index(drop=True)

df = movies.loc[(1980 <= movies["year"]) & (movies["year"] < 2000)].groupby("genre")["gross_earning"].nlargest(3)
max_genres  = df.groupby("genre").sum().sort_values()[-3:].index
max_indexes = df[max_genres].reset_index()["level_1"]
top_three_genres_Q22 = movies.loc[max_indexes,:].sort_values(sort_cols, ascending = [True, False]).reset_index(drop=True)

df = movies.loc[movies["year"] < 1980].groupby("genre")["gross_earning"].nlargest(3)
max_genres  = df.groupby("genre").sum().sort_values()[-3:].index
max_indexes = df[max_genres].reset_index()["level_1"]
top_three_genres_Q23 = movies.loc[max_indexes,:].sort_values(sort_cols, ascending = [True, False]).reset_index(drop=True)
````
