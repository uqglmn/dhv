### Question 1

The text file https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/files.txt contains a list of CSV files. <br>
Read all CSV files listed in the above file into a single pandas DataFrame ``grades``.

Then sort this DataFrame with respect to its **ID** column and reset the index.

DataFrames ``grades`` should then have the following head:

(see below)

**Points:** 1 point for the correct DataFrame ``grades``. 

**Important:** Your code will be tested on a different list of CSV files. Make sure your code reads the files and creates the DataFrame automatically.

Write your python code that completes this task in the textbox below, in plain text. <br>
There should be no import and print statements in your code. Do not write any comments.

````python
files = pd.read_csv("https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/files.txt", header=None)

grades = None
for i in files[0].values:
    df = pd.read_csv(i)
    grades = pd.concat([grades,df])
grades.sort_values("ID", inplace=True)
grades.reset_index(drop=True, inplace=True)

grades.head()
````

### Question 4

The CSV file https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/levels.csv contains additional information about columns of DataFrame ``grades``, except the **Group** and **ID** columns. <br>

Read this CSV file into a single pandas DataFrame ``levels``. Set the **Group** and **ID** columns of DataFrame ``grades`` to row MultiIndex. <br>
Then use DataFrame ``levels`` to construct column MultiIndex for DataFrame ``grades``.

DataFrames ``grades`` should then have the following head:

(see below)

**Points:** 1 points for the correct DataFrame ``grades`` after setting MultiIndexes.

**Hint:** In the last step, create dictionary ``levels_dict`` from DataFrame ``levels``. Then use this dictionary to create a list of tuples of level 0 and level 1 column names.
Finally, use ``pd.MultiIndex.from_tuples()`` to create the wanted column MultiIndex. 

Write your python code that completes this task in the textbox below, in plain text. <br>
There should be no import and print statements in your code. Do not write any comments.

````python
levels = pd.read_csv("https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/levels.csv")
levels_dict = dict(zip(levels['Level 1'], levels['Level 0']))

grades.set_index(["Group","ID"], inplace=True)
grades.sort_index(inplace = True)

pairs = [(levels_dict[i], i) for i in grades.columns]
grades.columns = pd.MultiIndex.from_tuples(pairs)

grades.head()
````

### Question 5

Compute mean **Level 4**, **Level 5**, **Level 6**, **Final Year** grades, and the overall mean grade (the **Final Grade**) for each student. Round the means to 1 decimal place precision.

Then compute Degree Class for each student using the following **Final Grade** classifier (use pd.cut):

|       |       |              |              |       |
|-------|-------|--------------|--------------|-------|
| **Final Grade** | (39.5, 49.5] | (49.5, 59.5] | (59.5, 69.5] | (69.5, 100] |
| **Class** | Third | Lower Second | Upper Second | First |

Lastly, combine all this data into DataFrame ``mean_grades``.

DataFrames ``mean_grades`` should have the following head:

(see below)

**Points:** 1 point for the correct DataFrame ``mean_grades``. 

Write your python code that completes this task in the textbox below, in plain text. <br>
There should be no import and print statements in your code. Do not write any comments.


````python
mean_grades = grades.iloc[:,:2]
mean_grades[("Annual Grades","Level 4")] = grades.loc[:,"Level 4"].mean(axis = 1).round(1)
mean_grades[("Annual Grades","Level 5")] = grades.loc[:,"Level 5"].mean(axis = 1).round(1)
mean_grades[("Annual Grades","Level 6")] = grades.loc[:,"Level 6"].mean(axis = 1).round(1)
mean_grades[("Annual Grades","Final Year")] = grades.loc[:,("Final Year","Project")].round(1)
mean_grades[("Degree","Final Grade")] = grades.loc[:,["Level 4","Level 5",
                                                      "Level 6","Final Year"]].mean(axis = 1).round(1)

classes = pd.cut(mean_grades[("Degree","Final Grade")],[39.50,49.50,59.50,69.50,100], 
                 labels=["Third", "Lower Second", "Upper Second", "First"])
mean_grades[("Degree","Class")] = classes

mean_grades.head()
````

### Question 6

Count the number of each **Degree Class** in DataFrame ``mean_grades``. <br>
Save your result to DataFrame ``degrees``. Then set column **Degree** to index.

DataFrames ``degrees`` should then have the following form:

(see below)

Save DataFrame ``degrees`` to the CSV file **"degrees.csv"**. The file contents should have the following form:

```
Degree,Count
Third,0
Lower Second,11
Upper Second,44
First,5
```

**Points:** 0.75 point for the correct DataFrame ``degrees``, and 0.25 point for the correct **"degrees.csv"** file.

Write your python code that completes this task in the textbox below, in plain text. <br>
There should be no import and print statements in your code. Do not write any comments.


````python
degrees = pd.DataFrame(mean_grades.pivot_table(index = [("Degree","Class")], aggfunc ='size'))
degrees.index.name = "Degree"
degrees.columns = ["Count"]

degrees.to_csv("degrees.csv")
degrees
````
